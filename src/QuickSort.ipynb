```c++
#include <iostream>
using namespace std;
```

```c++
template<class It>
void quicksort(It first, It last) {
    if(first == last)
        return;
        
    auto pivot = *next(first, distance(first, last)/2);
    
    It middle = partition(first, last, [pivot](auto x) { return x > pivot; });
    quicksort(first, middle);
    quicksort(next(middle), last);
}
```

```c++
vector<int> ints = {12, 23, 45, 24, 49, 9, 6};

quicksort(ints.begin(), ints.end());

for(auto i : ints)
    cout << i << " ";
```
